**Features ve labelları belirleme**

```python
features=penguin_df[['island','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','sex','year']]
label=penguin_df['species']
```

```python
display(features.head())
```

```python
display(label.head())
```

**Verisetindeki her sütunda bulunan eksik değer sayısını yazdırıyoruz.**

```python
missing_values=penguin_df.isnull().sum()
print(missing_values)
```

**Eksik sayısal değerleri ortalamayla, kategorik değişkenleri modla dolduruyoruz**

```python
features['bill_length_mm'] = features['bill_length_mm'].fillna(features['bill_length_mm'].mean())
features['bill_depth_mm'] = features['bill_depth_mm'].fillna(features['bill_depth_mm'].mean())
features['flipper_length_mm'] = features['flipper_length_mm'].fillna(features['flipper_length_mm'].mean())
features['body_mass_g'] = features['body_mass_g'].fillna(features['body_mass_g'].mean())

features['sex'] = features['sex'].fillna(features['sex'].mode()[0])
```

**Doldurduktan sonra, eksik değerleri tekrar kontrol ettik**

```python
missing_values=features.isnull().sum()
print(missing_values)
```

**Sex özelliğini işlerken sayısal değere dönüştürüyoruz**

```python
features['sex']=features['sex'].map({'male': 0, 'female': 1})
display(features['sex'].head())
```

**Aynı işleme get_dummies kullanarak da yapabiliriz**

```python
print(pd.get_dummies(features['sex'],dtype='int'))
```

**Verilerimizi işlemek için train-test olarak ayırıyoruz. X(bağımsız değişken) Y(bağımlı değişken)**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=42)

print("\nEğitim ve Test Seti Boyutları:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```
